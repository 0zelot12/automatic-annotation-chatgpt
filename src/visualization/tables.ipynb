```python
import latextable

from texttable import Texttable
```

```python
table_1 = Texttable()

table_1.set_cols_align(["l", "r", "r",])
table_1.set_cols_valign(["t", "m", "m"])

table_1.add_rows(entity_rows)

print(
    latextable.draw_latex(
        table_1,
        caption="Vorkommen von Entit채ten im PET-Datensatz",
        label="table:example_table",
    )
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
categories = ['Zero-Shot', 'One-Shot', 'Multi-Shot']
values1 = [0.01, 0.1, 0.2]
values2 = [0.04, 0.04, 0.09]
values3 = [0.1, 0.2, 0.4]

# Width of the bars
bar_width = 0.25

# Positions of bars on x-axis
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar chart
plt.bar(r1, values1, color='b', width=bar_width, label='F1-Score Gesamt')
plt.bar(r2, values2, color='g', width=bar_width, label='F1-Score Entit채ten')
plt.bar(r3, values3, color='r', width=bar_width, label='F1-Score Relationen')

# Add xticks on the middle of the group bars
plt.xlabel('Prompttyp')
plt.xticks([r + bar_width for r in range(len(categories))], categories)

plt.ylim((0, 1.0))

# Add labels and title
plt.ylabel('F1-Score')
plt.title('F1-Score Zero-Shot vs. One-Shot vs. Multi-Shot')

# Add legend
plt.legend()

# Show plot
plt.show()
```
