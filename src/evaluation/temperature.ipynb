```python
import os
import json

import numpy as np
```

```python
data_path = "../../data/results/low-temperature/5x5-relation-with-errors/"

data = []
for parent_folder in os.listdir(f"./{data_path}"):
    for child_folder in os.listdir(f"./{data_path}/{parent_folder}"):
        for file_name in os.listdir(f"./{data_path}/{parent_folder}/{child_folder}"):
            filepath = os.path.join(f"./{data_path}/{parent_folder}/{child_folder}", file_name)
            with open(filepath, "r") as file:
                json_data = json.load(file)
                data.append(json_data)

f1_scores_total_low_temperature = [d["metrics"]["overall_metrics"]["f1_score"] for d in data]
f1_scores_entity_low_temperature = [d["metrics"]["entity_metrics"]["overall"]["f1_score"] for d in data]
f1_scores_relation_low_temperature = [d["metrics"]["relation_metrics"]["overall"]["f1_score"] for d in data]

errors_low_temperature = np.sum([d["errors"] for d in data])

errors_low_temperature
```

```python
data_path = "../../data/results/two-shot/5x5-relation-with-errors/"

data = []
for parent_folder in os.listdir(f"./{data_path}"):
    for child_folder in os.listdir(f"./{data_path}/{parent_folder}"):
        for file_name in os.listdir(f"./{data_path}/{parent_folder}/{child_folder}"):
            filepath = os.path.join(f"./{data_path}/{parent_folder}/{child_folder}", file_name)
            with open(filepath, "r") as file:
                json_data = json.load(file)
                data.append(json_data)

f1_scores_total_medium_temperature = [d["metrics"]["overall_metrics"]["f1_score"] for d in data]
f1_scores_entity_medium_temperature = [d["metrics"]["entity_metrics"]["overall"]["f1_score"] for d in data]
f1_scores_relation_medium_temperature = [d["metrics"]["relation_metrics"]["overall"]["f1_score"] for d in data]

errors_medium_temperature = np.sum([d["errors"] for d in data])

errors_medium_temperature
```

```python
data_path = "../../data/results/high-temperature/5x5-relation-with-errors/"

data = []
for parent_folder in os.listdir(f"./{data_path}"):
    for child_folder in os.listdir(f"./{data_path}/{parent_folder}"):
        for file_name in os.listdir(f"./{data_path}/{parent_folder}/{child_folder}"):
            filepath = os.path.join(f"./{data_path}/{parent_folder}/{child_folder}", file_name)
            with open(filepath, "r") as file:
                json_data = json.load(file)
                data.append(json_data)

f1_scores_total_high_temperature = [d["metrics"]["overall_metrics"]["f1_score"] for d in data]
f1_scores_entity_high_temperature = [d["metrics"]["entity_metrics"]["overall"]["f1_score"] for d in data]
f1_scores_relation_high_temperature = [d["metrics"]["relation_metrics"]["overall"]["f1_score"] for d in data]

errors_high_temperature = np.sum([d["errors"] for d in data])

errors_high_temperature
```

```python
import matplotlib.pyplot as plt

values = [errors_low_temperature, errors_medium_temperature, errors_high_temperature]

plt.figure(figsize=(10, 5))

plt.bar(["Niedrige Temperature", "Mittlere Temperature", "Hohe Temperature"], values, color="r", label='Anzahl der Fehler bei 225 Annotationen')

plt.ylabel('#Fehler')
plt.title('Fehler Niedrige Temeperature vs. Mittlere Temperature vs. Hohe Temperature')

plt.legend()

plt.savefig("./images/vergleich_temperature_fehler.png", dpi=700)

plt.show()
```

```python
import matplotlib.pyplot as plt

values = [np.mean(f1_scores_total_low_temperature), np.mean(f1_scores_total_medium_temperature), np.mean(f1_scores_total_high_temperature)]

plt.figure(figsize=(10, 5))

plt.bar(["Niedrige Temperature", "Mittlere Temperature", "Hohe Temperature"], values, label='F1-Score bei 225 Annotationen')

plt.ylabel('F1-Score')
plt.title('F1-Score Temperature vs. Mittlere Temperature vs. Hohe Temperature')

plt.legend()

plt.ylim((0, 0.6))

plt.savefig("./images/vergleich_temperature_score.png", dpi=700)

plt.show()
```
